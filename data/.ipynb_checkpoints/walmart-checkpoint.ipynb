```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Optional: nicer plots
sns.set(style="whitegrid")
```

```python
file_path = '/Users/ashwinichintala/Desktop/walmart/data/walmart_cleaned_data.csv'
df = pd.read_csv(file_path)

# Check data
df.head()
df.info()
```

```python
# Total revenue per row
df['total'] = df['unit_price'] * df['quantity']

# Convert date to datetime
df['date'] = pd.to_datetime(df['date'], errors='coerce')

# Extract year
df['year'] = df['date'].dt.year
```

```python
revenue = df.groupby(['Branch','year'])['total'].sum().reset_index()
revenue.head()
```

```python
# Pivot for comparison
rev_pivot = revenue.pivot(index='Branch', columns='year', values='total').reset_index()

# Revenue decrease %
rev_pivot['rev_dec_ratio'] = ((rev_pivot[2022] - rev_pivot[2023]) / rev_pivot[2022]) * 100

# Top 5
top5 = rev_pivot.sort_values('rev_dec_ratio', ascending=False).head(5)
top5
```

```python
# Pivot for comparison
rev_pivot = revenue.pivot(index='Branch', columns='year', values='total').reset_index()

# Revenue decrease %
rev_pivot['rev_dec_ratio'] = ((rev_pivot[2022] - rev_pivot[2023]) / rev_pivot[2022]) * 100

# Top 5
top5 = rev_pivot.sort_values('rev_dec_ratio', ascending=False).head(5)
top5
```

```python
# Bar chart – Top 5 revenue decrease
```

```python
plt.figure(figsize=(10,6))
sns.barplot(data=top5, x='Branch', y='rev_dec_ratio', palette='Reds_r')
plt.title('Top 5 Branches with Highest Revenue Decrease (2023 vs 2022)')
plt.ylabel('Revenue Decrease (%)')
plt.xlabel('Branch')
plt.show()
```

```python
# Line Chart – Revenue Trend of Top 5 Branches (2022 vs 2023)
```

```python

top_branches = revenue.groupby('Branch')['total'].sum().nlargest(5).index
revenue_top = revenue[revenue['Branch'].isin(top_branches)]

plt.figure(figsize=(10,6))
sns.lineplot(data=revenue_top, x='year', y='total', hue='Branch', marker='o')
plt.title('Revenue Trend of Top 5 Branches')
plt.ylabel('Revenue')
plt.xlabel('Year')
plt.show()
```

```python
import pandas as pd

# Convert 'date' to datetime type
df['date'] = pd.to_datetime(df['date'], errors='coerce')

# Extract year and month
df['year'] = df['date'].dt.year
df['month'] = df['date'].dt.month

# Create a total revenue column
df['total'] = df['unit_price'] * df['quantity']

# Check if it worked
df[['date', 'year', 'month', 'total']].head()
```

```python
# Create a revenue summary per branch per year
```

```python
revenue = df.groupby(['Branch', 'year'], as_index=False)['total'].sum()
revenue.head()
```

```python
# Visualize Top 5 Branch Revenue Trend
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Identify top 5 branches
top_branches = revenue.groupby('Branch')['total'].sum().nlargest(5).index
revenue_top = revenue[revenue['Branch'].isin(top_branches)]

# Plot line chart
plt.figure(figsize=(10,6))
sns.lineplot(data=revenue_top, x='year', y='total', hue='Branch', marker='o')
plt.title('Revenue Trend of Top 5 Branches')
plt.ylabel('Revenue')
plt.xlabel('Year')
plt.show()
```

```python
# Make sure to have the 2022 and 2023 revenue data
```

```python
# Revenue for 2022
revenue_2022 = df[df['year'] == 2022].groupby('Branch', as_index=False)['total'].sum()
revenue_2022.rename(columns={'total': 'rev_2022'}, inplace=True)

# Revenue for 2023
revenue_2023 = df[df['year'] == 2023].groupby('Branch', as_index=False)['total'].sum()
revenue_2023.rename(columns={'total': 'rev_2023'}, inplace=True)
```

```python
# Merge to calculate revenue decrease ratio
```

```python
top5 = revenue_2022.merge(revenue_2023, on='Branch')
top5['rev_dec_ratio'] = ((top5['rev_2022'] - top5['rev_2023']) / top5['rev_2022']) * 100

# Keep only branches where revenue decreased
top5 = top5[top5['rev_dec_ratio'] > 0]

# Sort by highest decrease and take top 5
top5 = top5.sort_values(by='rev_dec_ratio', ascending=False).head(5)
```

```python
# Now plot the bar chart
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10,6))
sns.barplot(data=top5, x='Branch', y='rev_dec_ratio', palette='Reds_r')
plt.title('Top 5 Branches with Highest Revenue Decrease (2023 vs 2022)')
plt.ylabel('Revenue Decrease (%)')
plt.xlabel('Branch')

# Optional: save the chart
plt.savefig('/Users/ashwinichintala/Desktop/top5_revenue_decrease.png', dpi=300, bbox_inches='tight')

plt.show()
```
